```python
%matplotlib notebook
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
# %matplotlib inline
```

```python
Y = np.loadtxt('tr5000_N800/printY/Y_800_0.001_1.800.txt')
target = np.loadtxt('tr5000_N800/printY/T_800_0.001_1.800.txt')
```

```python
th_min = -3.1972
th_scale = 5.6622
```

```python
np.max(Y)
```

```python
sep_idx = np.where(target==1)[0]
healthy_idx = np.where(target==0)[0]
```

```python
Y_sep_800 = (Y[sep_idx] - th_min)/th_scale
Y_healthy_800 = (Y[healthy_idx] - th_min)/th_scale
```

```python
plt.figure()
plt.hist(Y_sep_800, bins=500, color='red', alpha=0.5);
plt.hist(Y_healthy_800, bins=500, color='blue', alpha=0.5);

```
